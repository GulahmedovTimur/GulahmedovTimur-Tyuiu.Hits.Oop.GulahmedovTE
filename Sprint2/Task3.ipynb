#### Дополнительное задание
Добавьте к сущестующим классам (базовыму и производным 3-4 атрибута и метода) и реализуйте простое, сложное и множественное наследование

```
using System;
using System.Collections.Generic;

// Базовый класс для всех продуктов
public class Product
{
    private string name;
    private decimal price;
    private int quantity;
    protected string category;
    internal string description;
    private string manufacturer;
    private DateTime productionDate;
    protected bool isEcoFriendly;
    internal int warrantyMonths;

    public Product(string name, decimal price, int quantity, string category, string description,
                   string manufacturer, DateTime productionDate, bool isEcoFriendly, int warrantyMonths)
    {
        this.name = name;
        this.price = price;
        this.quantity = quantity;
        this.category = category;
        this.description = description;
        this.manufacturer = manufacturer;
        this.productionDate = productionDate;
        this.isEcoFriendly = isEcoFriendly;
        this.warrantyMonths = warrantyMonths;
    }

    public string Name { get => name; set => name = value; }
    public decimal Price { get => price; set => price = value; }
    public int Quantity { get => quantity; set => quantity = value; }
    public string Category { get => category; protected set => category = value; }
    public string Description { get => description; internal set => description = value; }
    public string Manufacturer { get => manufacturer; set => manufacturer = value; }
    public DateTime ProductionDate { get => productionDate; set => productionDate = value; }
    public bool IsEcoFriendly { get => isEcoFriendly; protected set => isEcoFriendly = value; }
    public int WarrantyMonths { get => warrantyMonths; internal set => warrantyMonths = value; }

    public decimal GetTotalPrice() => price * quantity;
    
    public void UpdateQuantity(int amount)
    {
        quantity += amount;
        if (quantity < 0) quantity = 0;
    }

    public string GetProductInfo()
    {
        return $"Название: {name}, Цена: {price}, Количество: {quantity}, Категория: {category}";
    }

    // Новые методы
    public bool IsExpired()
    {
        return productionDate.AddYears(2) < DateTime.Now;
    }

    public virtual void ApplyDiscount(decimal percent)
    {
        price -= price * (percent / 100);
        Console.WriteLine($"Скидка {percent}% применена. Новая цена: {price}");
    }

    public string GetManufacturerInfo()
    {
        return $"Производитель: {manufacturer}, Дата производства: {productionDate.ToShortDateString()}";
    }

    public void ExtendWarranty(int additionalMonths)
    {
        warrantyMonths += additionalMonths;
        Console.WriteLine($"Гарантия продлена на {additionalMonths} месяцев");
    }
}

// ПРОСТОЕ НАСЛЕДОВАНИЕ (один родительский класс)
public class ElectronicProduct : Product
{
    // Новые атрибуты
    private int powerConsumption;
    private string energyClass;

    public ElectronicProduct(string name, decimal price, int quantity, string description,
                            string manufacturer, DateTime productionDate, bool isEcoFriendly,
                            int warrantyMonths, int powerConsumption, string energyClass)
        : base(name, price, quantity, "Электроника", description, manufacturer, 
              productionDate, isEcoFriendly, warrantyMonths)
    {
        this.powerConsumption = powerConsumption;
        this.energyClass = energyClass;
    }

    // Новые методы
    public double CalculateYearlyEnergyCost(double kWhPrice)
    {
        return (powerConsumption * 365 * 8 / 1000.0) * kWhPrice; // 8 часов в день
    }

    public void UpdateEnergyClass(string newClass)
    {
        energyClass = newClass;
        Console.WriteLine($"Энергокласс обновлен: {energyClass}");
    }

    public override void ApplyDiscount(decimal percent)
    {
        base.ApplyDiscount(percent);
        Console.WriteLine("Скидка на электротехнику!");
    }

    public string GetEnergyInfo()
    {
        return $"Потребляемая мощность: {powerConsumption}Вт, Энергокласс: {energyClass}";
    }
}

// СЛОЖНОЕ НАСЛЕДОВАНИЕ (цепочка наследования)
public class Smartphone : ElectronicProduct
{

    private string os;
    private int cameraMP;

    public Smartphone(string name, decimal price, int quantity, string description,
                     string manufacturer, DateTime productionDate, bool isEcoFriendly,
                     int warrantyMonths, int powerConsumption, string energyClass,
                     string os, int cameraMP)
        : base(name, price, quantity, description, manufacturer, productionDate,
              isEcoFriendly, warrantyMonths, powerConsumption, energyClass)
    {
        this.os = os;
        this.cameraMP = cameraMP;
    }

    // Новые методы
    public void InstallUpdate()
    {
        Console.WriteLine($"Установка обновления для {os}");
    }

    public void TakePhoto()
    {
        Console.WriteLine($"Сделано фото на {cameraMP}МП");
    }

    public override void ApplyDiscount(decimal percent)
    {
        base.ApplyDiscount(percent);
        Console.WriteLine("Дополнительная скидка на смартфон!");
    }

    public string GetTechSpecs()
    {
        return $"ОС: {os}, Камера: {cameraMP}МП, {GetEnergyInfo()}";
    }
}

// ИНТЕРФЕЙСЫ для множественного наследования
public interface ISellable
{
    void MarkAsSold();
    bool IsAvailable { get; }
}

public interface IDiscountable
{
    void ApplySeasonalDiscount();
}

// МНОЖЕСТВЕННОЕ НАСЛЕДОВАНИЕ через интерфейсы
public class Book : Product, ISellable, IDiscountable
{
    private string author;
    private int pageCount;

    public Book(string name, decimal price, int quantity, string description,
               string manufacturer, DateTime productionDate, bool isEcoFriendly,
               int warrantyMonths, string author, int pageCount)
        : base(name, price, quantity, "Книги", description, manufacturer,
              productionDate, isEcoFriendly, warrantyMonths)
    {
        this.author = author;
        this.pageCount = pageCount;
    }

    // Реализация интерфейса
    public void MarkAsSold()
    {
        Quantity = 0;
        Console.WriteLine($"Книга '{Name}' помечена как проданная");
    }

    public bool IsAvailable => Quantity > 0;

    // Реализация интерфейса
    public void ApplySeasonalDiscount()
    {
        ApplyDiscount(15);
        Console.WriteLine("Сезонная скидка на книгу применена!");
    }

    // Новые методы
    public string GetAuthorInfo()
    {
        return $"Автор: {author}, Страниц: {pageCount}";
    }

    public void ReadSample()
    {
        Console.WriteLine($"Чтение образца книги '{Name}'...");
    }
}

public class Seller
{
    private string name;
    private string employeeId;
    protected decimal salary;
    internal string contactInfo;
    private List<Product> products = new List<Product>();
    private int salesCount;
    private DateTime hireDate;
    protected string department;

    public Seller(string name, string employeeId, decimal salary, string contactInfo,
                 DateTime hireDate, string department)
    {
        this.name = name;
        this.employeeId = employeeId;
        this.salary = salary;
        this.contactInfo = contactInfo;
        this.hireDate = hireDate;
        this.department = department;
        this.salesCount = 0;
    }

    public string Name { get => name; set => name = value; }
    public string EmployeeId { get => employeeId; set => employeeId = value; }
    public decimal Salary { get => salary; set => salary = value; }
    public string ContactInfo { get => contactInfo; set => contactInfo = value; }
    public int SalesCount { get => salesCount; private set => salesCount = value; }
    public DateTime HireDate { get => hireDate; set => hireDate = value; }
    public string Department { get => department; protected set => department = value; }

    public void AddProduct(Product product) => products.Add(product);

    public List<Product> GetProducts()
    {
        return products;
    }

    public void SellProduct(Product product, int quantity)
    {
        if (product.Quantity >= quantity)
        {
            product.UpdateQuantity(-quantity);
            salesCount++;
            Console.WriteLine($"Продано {quantity} единиц товара {product.Name}");
        }
        else
        {
            Console.WriteLine($"Недостаточно товара {product.Name} на складе");
        }
    }

    // Новые методы
    public void IncreaseSalary(decimal amount)
    {
        salary += amount;
        Console.WriteLine($"Зарплата увеличена на {amount}. Новая зарплата: {salary}");
    }

    public int CalculateWorkExperience()
    {
        return (DateTime.Now - hireDate).Days / 365;
    }

    public string GetSalesStats()
    {
        return $"Продавец {name}: {salesCount} продаж, опыт работы: {CalculateWorkExperience()} лет";
    }

    public void ChangeDepartment(string newDepartment)
    {
        department = newDepartment;
        Console.WriteLine($"Продавец переведен в отдел: {department}");
    }

    public string GetSellerInfo()
    {
        return $"Имя: {name}, ID: {employeeId}, Зарплата: {salary}, Контакты: {contactInfo}";
    }
}

public class Store
{
    private string storeName;
    private string location;
    public string storeHours;
    private decimal revenue;
    private int customerCount;
    private List<Seller> sellers = new List<Seller>();

    public Store(string storeName, string location, string storeHours)
    {
        this.storeName = storeName;
        this.location = location;
        this.storeHours = storeHours;
        this.revenue = 0;
        this.customerCount = 0;
    }

    public string StoreName { get => storeName; set => storeName = value; }
    public string Location { get => location; set => location = value; }
    public string StoreHours { get => storeHours; set => storeHours = value; }
    public decimal Revenue { get => revenue; private set => revenue = value; }
    public int CustomerCount { get => customerCount; private set => customerCount = value; }


    public void AddSeller(Seller seller) => sellers.Add(seller);

    public void ListProducts()
    {
        Console.WriteLine($"\nТовары в магазине {storeName}:");
        foreach (var seller in sellers)
        {
            foreach (var product in seller.GetProducts())
            {
                Console.WriteLine(product.GetProductInfo());
            }
        }
    }

    // Новые методы
    public void AddRevenue(decimal amount)
    {
        revenue += amount;
        customerCount++;
        Console.WriteLine($"Выручка увеличена на {amount}. Общая выручка: {revenue}");
    }

    public void PrintStoreStats()
    {
        Console.WriteLine($"\nСтатистика магазина {storeName}:");
        Console.WriteLine($"Выручка: {revenue}");
        Console.WriteLine($"Количество покупателей: {customerCount}");
        Console.WriteLine($"Количество продавцов: {sellers.Count}");
    }

    public void Open()
    {
        Console.WriteLine($"Магазин {storeName} открыт! Время работы: {storeHours}");
    }

    public void Close()
    {
        Console.WriteLine($"Магазин {storeName} закрыт!");
    }

    public string GetStoreInfo()
    {
        return $"Магазин: {storeName}, Адрес: {location}, Время работы: {storeHours}";
    }
}


Console.WriteLine("=== ДЕМОНСТРАЦИЯ СИСТЕМЫ МАГАЗИНА ===\n");

// Создание продуктов разных типов
var smartphone = new Smartphone(
    "iPhone 15", 150000, 10, "Флагманский смартфон",
    "Apple", new DateTime(2023, 1, 1), true, 12, 20, "A+",
    "iOS", 48
);

var book = new Book(
    "Война и мир", 1500, 5, "Классическая литература",
    "Эксмо", new DateTime(2020, 1, 1), true, 0,
    "Лев Толстой", 1225
);

// Создание продавца
var seller = new Seller(
    "Иван Петров", "S001", 50000, "ivan@store.ru",
    new DateTime(2020, 5, 1), "Электроника"
);

// Создание магазина
var store = new Store("ТехноМир", "ул. Пушкина, 1", "10:00-22:00");

// Добавление продавца в магазин
store.AddSeller(seller);

// Добавление товаров продавцу
seller.AddProduct(smartphone);
seller.AddProduct(book);

store.Open();
store.ListProducts();

Console.WriteLine("\n=== ДЕМОНСТРАЦИЯ НОВЫХ МЕТОДОВ ===");

// Демонстрация новых методов
smartphone.ApplyDiscount(10);
book.ApplySeasonalDiscount();
seller.IncreaseSalary(5000);

Console.WriteLine("\n=== ПРОЦЕСС ПРОДАЖИ ===");
seller.SellProduct(smartphone, 2);
store.AddRevenue(270000);

Console.WriteLine("\n=== РАБОТА С ИНТЕРФЕЙСАМИ ===");
Console.WriteLine($"Книга доступна до продажи: {book.IsAvailable}");
book.MarkAsSold();
Console.WriteLine($"Книга доступна после продажи: {book.IsAvailable}");

Console.WriteLine("\n=== ДЕМОНСТРАЦИЯ НАСЛЕДОВАНИЯ ===");
Console.WriteLine("Простое наследование: " + smartphone.GetEnergyInfo());
Console.WriteLine("Сложное наследование: " + smartphone.GetTechSpecs());
Console.WriteLine("Множественное наследование: " + book.GetAuthorInfo());

Console.WriteLine("\n=== СТАТИСТИКА ===");
store.PrintStoreStats();
Console.WriteLine(seller.GetSalesStats());

store.Close();
```
